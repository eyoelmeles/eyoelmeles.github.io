# 实验1.1 线性基函数回归实验

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec  # 用于更精细的子图布局
```

```python
# --- 1. 数据生成函数 ---
def generate_data(num_train_samples, num_test_samples, noise_std=0.5):
    """
    生成sin函数+高斯噪声的数据集。
    训练数据在 [-5, 5] 范围内生成，测试数据在 [-10, 10] 范围内生成。
    """
    # 训练数据
    x_train = np.random.uniform(-5, 5, num_train_samples)
    x_train = np.sort(x_train)  # 排序以便绘制连续曲线
    y_train = 2 * np.sin(0.5 * x_train) + np.random.normal(0, noise_std, num_train_samples)

    # 测试数据
    x_test = np.random.uniform(-10, 10, num_test_samples)
    x_test = np.sort(x_test)  # 排序以便绘制连续曲线
    y_test = 2 * np.sin(0.5 * x_test) + np.random.normal(0, noise_std, num_test_samples)

    # 真实的sin函数曲线（用于绘图对比）
    x_true = np.linspace(-10, 10, 500)
    y_true = 2 * np.sin(0.5 * x_true)

    return x_train, y_train, x_test, y_test, x_true, y_true
```

```python
# --- 2. 基函数定义 ---
def polynomial_basis_function(x, K):
    """
    多项式基函数。
    Φ(x) = [x^0, x^1, ..., x^K]。
    K 代表多项式的最高次幂，因此会生成 K+1 个基函数。
    """
    # 确保x是numpy数组，以便进行广播操作
    if not isinstance(x, np.ndarray):
        x = np.array(x)

    # 构建特征矩阵，每一列是x的幂
    # 例如，如果x是[x1, x2]且K=2，则features将是
    # [[x1^0, x1^1, x1^2],
    #  [x2^0, x2^1, x2^2]]
    features = np.array([x ** k for k in range(K + 1)]).T
    return features
```

```python
# --- 3. 线性基函数回归模型实现 ---
def train_linear_regression(Phi, y):
    """
    使用极大似然原理计算w_MLE。
    w_MLE = (Phi.T @ Phi + lambda_reg * I)^-1 @ Phi.T @ y
    添加一个小的正则化项 lambda_reg * I 来避免奇异矩阵问题。
    """
    lambda_reg = 1e-6  # 正则化参数，防止 Phi.T @ Phi 不可逆
    identity_matrix = np.identity(Phi.shape[1])  # 创建与特征数量匹配的单位矩阵

    # 计算 (Phi.T @ Phi + lambda_reg * I) 的逆
    # @ 运算符在Python中表示矩阵乘法
    try:
        w_MLE = np.linalg.inv(Phi.T @ Phi + lambda_reg * identity_matrix) @ Phi.T @ y
    except np.linalg.LinAlgError:
        print("Warning: Matrix is singular. Cannot compute inverse for training.")
        return None  # 或者处理错误

    return w_MLE
```

```python
def predict(Phi, w):
    """
    使用训练好的w进行预测。
    y_pred = Phi @ w
    """
    return Phi @ w
```

```python
def calculate_rmse(y_true, y_pred):
    """
    计算均方根误差 (RMSE)。
    RMSE = sqrt(mean((y_true - y_pred)^2))
    """
    return np.sqrt(np.mean((y_true - y_pred) ** 2))
```

```python
# --- 4. 实验配置 ---
train_data_configs = {
    'S': {'num_train': 10, 'num_test': 20},  # Small dataset
    'M': {'num_train': 30, 'num_test': 40},  # Medium dataset
    'L': {'num_train': 60, 'num_test': 80}  # Large dataset
}
K_values = [2, 4, 8]  # K阶多项式，表示最高次幂，实际基函数数量为K+1

results = {}  # 存储所有实验的数值结果（w_MLE, RMSE, y_pred_curve等）
```

```python
# --- 5. 绘图美化设置 ---
# 使用matplotlib默认样式，通常背景为白色
plt.style.use('default')
plt.rcParams.update({
    'font.size': 10,  # 整体字体大小
    'axes.labelsize': 10,  # 坐标轴标签字体大小
    'axes.titlesize': 12,  # 标题字体大小
    'xtick.labelsize': 8,  # x轴刻度标签字体大小
    'ytick.labelsize': 8,  # y轴刻度标签字体大小
    'legend.fontsize': 8,  # 图例字体大小
    'figure.titlesize': 14,  # 主图标题字体大小
    'lines.linewidth': 1.5,  # 曲线线宽
    'scatter.marker': 'o',  # 散点默认标记
    'axes.edgecolor': 'lightgray',  # 坐标轴边框颜色
    'axes.facecolor': 'white',  # 绘图区域背景色
    'figure.facecolor': 'white',  # 整个figure的背景色
    'grid.color': 'lightgray',  # 网格颜色
    'grid.linestyle': '--',  # 网格线型
    'grid.linewidth': 0.5  # 网格线宽
})
```

```python
# --- 6. 创建Figure并进行循环绘图 ---
# 创建一个更大的Figure来容纳所有子图
fig = plt.figure(figsize=(15, 12), dpi=100, constrained_layout=True)  # 明确创建Figure对象，并调整尺寸和DPI

# 使用gridspec更灵活地控制子图间距，指定行数、列数、宽度间距和高度间距
gs = gridspec.GridSpec(len(K_values), len(train_data_configs), wspace=0.25, hspace=0.35)

# 循环不同的训练数据规模 (列)
for col_idx, data_label in enumerate(train_data_configs.keys()):
    num_train = train_data_configs[data_label]['num_train']
    num_test = train_data_configs[data_label]['num_test']

    # 生成当前规模的数据
    x_train, y_train, x_test, y_test, x_true, y_true = generate_data(num_train, num_test)

    # 循环不同的模型复杂度 K (行)
    for row_idx, K in enumerate(K_values):
        # 通过 fig.add_subplot() 将 Axes 对象添加到 Figure 中
        # 并通过 gs[row_idx, col_idx] 指定其在网格中的位置
        ax = fig.add_subplot(gs[row_idx, col_idx])

        # 训练模型
        Phi_train = polynomial_basis_function(x_train, K)
        w_MLE = train_linear_regression(Phi_train, y_train)

        if w_MLE is None:  # 处理训练失败的情况
            ax.set_title(f'K={K} Data={data_label} - Training Failed')
            ax.text(0.5, 0.5, 'Matrix Singular', horizontalalignment='center', verticalalignment='center',
                    transform=ax.transAxes)
            ax.set_xlim(-10, 10)
            ax.set_ylim(-4, 4)
            continue  # 跳过当前子图的绘图

        # 在测试数据上进行预测
        Phi_test = polynomial_basis_function(x_test, K)
        y_pred_test = predict(Phi_test, w_MLE)
        rmse = calculate_rmse(y_test, y_pred_test)

        # 预测曲线（在完整的x_true范围内）
        Phi_true_curve = polynomial_basis_function(x_true, K)
        y_pred_curve = predict(Phi_true_curve, w_MLE)

        # 存储结果 (可用于后续分析)
        results[f'K={K}_{data_label}'] = {
            'w_MLE': w_MLE,
            'rmse': rmse,
            'y_pred_curve': y_pred_curve,
            'x_train': x_train, 'y_train': y_train,
            'x_test': x_test, 'y_test': y_test
        }

        # --- 7. 绘图细节（所有绘图操作都通过 ax 对象进行） ---
        ax.scatter(x_test, y_test, color='#E74C3C', marker='o', s=15, label='Test Samples', alpha=0.7)  # 红色圆点
        ax.scatter(x_train, y_train, color='#34495E', marker='+', s=25, label='Train Samples')  # 黑色十字
        ax.plot(x_true, y_true, color='#2ECC71', linestyle='-', linewidth=1.5, label='True Sin Curve')  # 绿色实线
        ax.plot(x_true, y_pred_curve, color='#3498DB', linestyle='-', linewidth=1.5, label='MLE Curve')  # 蓝色实线

        ax.set_title(f'K={K} Data={data_label} RMSE: {rmse:.4f}', fontsize=10)  # 子图标题
        ax.set_xlabel('X', fontsize=9)  # x轴标签
        ax.set_ylabel('Y', fontsize=9)  # y轴标签

        # 图例美化，放置在右上角，有边框，无阴影
        ax.legend(loc='upper right', frameon=True, shadow=False, borderpad=0.5, fancybox=True)
        ax.grid(True, linestyle='--', alpha=0.6)  # 显示网格线，虚线，半透明
        ax.set_xlim(-10, 10)  # 设置x轴范围
        ax.set_ylim(-4, 4)  # 设置y轴范围

        # 调整刻度线的颜色和宽度
        ax.tick_params(axis='x', colors='gray', width=0.5)
        ax.tick_params(axis='y', colors='gray', width=0.5)
# --- 8. 添加主标题并调整布局 ---
fig.suptitle('Linear Basis Function Regression Experiment Results', fontsize=16)  # 为整个Figure添加主标题
#自动调整子图参数，使之填充整个 figure 区域，避免重叠
# --- 9. 显示图表 ---
plt.show()
```

# 实验1.1结果分析

```python
#这个实验清晰地展示了模型复杂度 (K) 和训练数据量对线性基函数回归性能的关键影响。为了获得最佳性能，需要在模型复杂度与数据量之间找到一个平衡点，以避免欠拟合和过拟合。对于本例中的正弦函数，一个中等复杂度的模型 (K=4) 配合足够的训练数据 (Data=M) 取得了最好的效果。过高或过低的模型复杂度，即便在不同数据量下，也难以达到理想的泛化性能。
```

# 实验1.2 多分类实验

```python
#加载Iris数据集并进行PCA降维以便可视化
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# 加载Iris数据集
iris = load_iris()
X_full = iris.data
y_full = iris.target
target_names = iris.target_names
feature_names = iris.feature_names

print("原始特征维度:", X_full.shape)
print("类别名称:", target_names)

# 数据标准化（对PCA和后续模型训练都有益）
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_full)

# PCA降维到2维以便可视化
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled) # 对标准化后的数据进行PCA

print("PCA降维后维度:", X_pca.shape)

# 可视化降维后的数据
plt.figure(figsize=(8, 6))
for i, target_name in enumerate(target_names):
    plt.scatter(X_pca[y_full == i, 0], X_pca[y_full == i, 1], label=target_name, alpha=0.8)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of IRIS dataset')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# 为了模型训练，我们通常还需要将数据分成训练集和测试集
# 这里我们直接使用PCA后的数据进行演示，实际应用中会在PCA之前划分
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y_full, test_size=0.3, random_state=42, stratify=y_full)

print(f"训练集PCA数据形状: {X_train_pca.shape}")
print(f"测试集PCA数据形状: {X_test_pca.shape}")
```

```python
# --- Logistic Regression 实现 ---

def sigmoid(z):
    """Sigmoid激活函数"""
    return 1 / (1 + np.exp(-z))

def compute_loss(X, y, w):
    """计算二分类交叉熵损失"""
    m = X.shape[0]
    h = sigmoid(X @ w)
    # 避免log(0)发生，对h进行裁剪
    h = np.clip(h, 1e-10, 1 - 1e-10)
    loss = (-1 / m) * np.sum(y * np.log(h) + (1 - y) * np.log(1 - h))
    return loss

def gradient_descent(X, y, learning_rate, num_iterations):
    """
    使用梯度下降训练Logistic Regression模型。
    X: 特征矩阵 (m, n)
    y: 标签向量 (m,)
    learning_rate: 学习率
    num_iterations: 迭代次数
    """
    m, n = X.shape
    # 初始化权重w，额外加一列作为偏置项
    w = np.zeros(n + 1)
    # 将偏置项x0=1添加到X
    X_b = np.c_[np.ones(m), X]

    losses = []

    for i in range(num_iterations):
        h = sigmoid(X_b @ w)
        gradient = (1 / m) * X_b.T @ (h - y)
        w -= learning_rate * gradient
        loss = compute_loss(X_b, y, w)
        losses.append(loss)

        if (i + 1) % 1000 == 0:
            print(f"Iteration {i+1}/{num_iterations}, Loss: {loss:.4f}")

    return w, losses

def predict_logistic_regression(X, w_trained, threshold=0.5):
    """
    使用训练好的权重进行预测。
    X: 特征矩阵
    w_trained: 训练好的权重 (包含偏置项)
    """
    m = X.shape[0]
    X_b = np.c_[np.ones(m), X] # 添加偏置项
    probabilities = sigmoid(X_b @ w_trained)
    return (probabilities >= threshold).astype(int)

# --- 可视化决策边界的辅助函数 ---
def plot_decision_boundary(ax, X, y, w, title, target_names, class_labels):
    """绘制Logistic Regression的决策边界"""
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                         np.arange(y_min, y_max, 0.02))

    # 使用训练好的权重预测网格点
    Z = predict_logistic_regression(np.c_[xx.ravel(), yy.ravel()], w)
    Z = Z.reshape(xx.shape)

    # 绘制背景色区域
    ax.contourf(xx, yy, Z, alpha=0.4, cmap=plt.cm.coolwarm)

    # 绘制散点图
    # 将y映射回0和1，以对应二分类器的标签
    y_plot = np.where(y == class_labels[1], 1, 0)
    ax.scatter(X[:, 0], X[:, 1], c=y_plot, cmap=plt.cm.coolwarm, edgecolor='k', s=20)

    ax.set_title(title)
    ax.set_xlabel('Principal Component 1')
    ax.set_ylabel('Principal Component 2')
    ax.set_xlim(xx.min(), xx.max())
    ax.set_ylim(yy.min(), yy.max())
    # 添加类别标签，确保颜色对应
    for i, label_idx in enumerate(class_labels):
        ax.scatter([], [], label=f'{target_names[label_idx]}',
                   color=plt.cm.coolwarm(i), edgecolor='k', s=20)
    ax.legend(loc='upper right')
    ax.grid(True, linestyle='--', alpha=0.6)
```

```python
# --- 训练和可视化三个二分类器 ---

learning_rate = 0.1
num_iterations = 10000

# 存储训练好的权重
w_0_vs_1, w_0_vs_2, w_1_vs_2 = None, None, None

# 0 vs 1 分类器 (setosa vs versicolor)
# 标签转换为0和1
X_01 = X_train_pca[(y_train == 0) | (y_train == 1)]
y_01 = y_train[(y_train == 0) | (y_train == 1)]
y_01_binary = np.where(y_01 == 1, 1, 0) # setosa为0, versicolor为1
w_0_vs_1, _ = gradient_descent(X_01, y_01_binary, learning_rate, num_iterations)
print(f"\nCompleted training for 0 vs 1: {w_0_vs_1}")

# 0 vs 2 分类器 (setosa vs virginica)
X_02 = X_train_pca[(y_train == 0) | (y_train == 2)]
y_02 = y_train[(y_train == 0) | (y_train == 2)]
y_02_binary = np.where(y_02 == 2, 1, 0) # setosa为0, virginica为1
w_0_vs_2, _ = gradient_descent(X_02, y_02_binary, learning_rate, num_iterations)
print(f"Completed training for 0 vs 2: {w_0_vs_2}")

# 1 vs 2 分类器 (versicolor vs virginica)
X_12 = X_train_pca[(y_train == 1) | (y_train == 2)]
y_12 = y_train[(y_train == 1) | (y_train == 2)]
y_12_binary = np.where(y_12 == 2, 1, 0) # versicolor为0, virginica为1
w_1_vs_2, _ = gradient_descent(X_12, y_12_binary, learning_rate, num_iterations)
print(f"Completed training for 1 vs 2: {w_1_vs_2}")


# --- 绘制三个二分类器的决策边界 ---
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

plot_decision_boundary(axes[0], X_train_pca[(y_train == 0) | (y_train == 1)], y_train[(y_train == 0) | (y_train == 1)],
                       w_0_vs_1, 'Decision Boundary: Class 0 vs Class 1', target_names, [0, 1])
plot_decision_boundary(axes[1], X_train_pca[(y_train == 0) | (y_train == 2)], y_train[(y_train == 0) | (y_train == 2)],
                       w_0_vs_2, 'Decision Boundary: Class 0 vs Class 2', target_names, [0, 2])
plot_decision_boundary(axes[2], X_train_pca[(y_train == 1) | (y_train == 2)], y_train[(y_train == 1) | (y_train == 2)],
                       w_1_vs_2, 'Decision Boundary: Class 1 vs Class 2', target_names, [1, 2])

plt.tight_layout()
plt.show()

# --- 投票法多分类预测 ---

def predict_one_vs_one_voting(X, w_01, w_02, w_12):
    """
    使用投票法进行多分类预测。
    X: 特征矩阵
    w_ij: 类别i vs 类别j 的二分类器权重 (包含偏置项)
    返回预测的类别标签
    """
    m = X.shape[0]
    predictions = np.zeros((m, 3)) # 存储每个分类器对每个类别的投票

    # 0 vs 1 分类器预测
    # 如果预测为0，则投给类别0；如果预测为1，则投给类别1
    pred_01 = predict_logistic_regression(X, w_01)
    predictions[pred_01 == 0, 0] += 1 # 投给0
    predictions[pred_01 == 1, 1] += 1 # 投给1

    # 0 vs 2 分类器预测
    # 如果预测为0，则投给类别0；如果预测为1（即原标签2），则投给类别2
    pred_02 = predict_logistic_regression(X, w_02)
    predictions[pred_02 == 0, 0] += 1 # 投给0
    predictions[pred_02 == 1, 2] += 1 # 投给2

    # 1 vs 2 分类器预测
    # 如果预测为0（即原标签1），则投给类别1；如果预测为1（即原标签2），则投给类别2
    pred_12 = predict_logistic_regression(X, w_12)
    predictions[pred_12 == 0, 1] += 1 # 投给1
    predictions[pred_12 == 1, 2] += 1 # 投给2

    # 找出投票最多的类别作为最终预测
    final_predictions = np.argmax(predictions, axis=1)
    return final_predictions

# 对测试集进行预测
y_pred_voting = predict_one_vs_one_voting(X_test_pca, w_0_vs_1, w_0_vs_2, w_1_vs_2)

# 计算准确率
accuracy_voting = np.mean(y_pred_voting == y_test)
print(f"\nVoting Classifier Test Accuracy: {accuracy_voting:.4f}")


# --- 绘制投票法多分类的决策边界 ---
plt.figure(figsize=(7, 6))

x_min, x_max = X_pca[:, 0].min() - 1, X_pca[:, 0].max() + 1
y_min, y_max = X_pca[:, 1].min() - 1, X_pca[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))

# 使用投票法预测网格点
Z_voting = predict_one_vs_one_voting(np.c_[xx.ravel(), yy.ravel()], w_0_vs_1, w_0_vs_2, w_1_vs_2)
Z_voting = Z_voting.reshape(xx.shape)

# 绘制背景色区域
plt.contourf(xx, yy, Z_voting, alpha=0.4, cmap=plt.cm.viridis) # 使用不同的颜色映射

# 绘制散点图
for i, target_name in enumerate(target_names):
    plt.scatter(X_pca[y_full == i, 0], X_pca[y_full == i, 1], label=target_name, alpha=0.8,
                cmap=plt.cm.viridis, edgecolor='k', s=20) # 原始数据点

plt.title(f'Multi-class Decision Boundary (Voting Classifier), Accuracy: {accuracy_voting:.4f}')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# --- Softmax Regression 实现 ---

def softmax(z):
    """Softmax激活函数"""
    # 避免数值溢出，对z进行稳定性处理
    exp_z = np.exp(z - np.max(z, axis=1, keepdims=True))
    return exp_z / np.sum(exp_z, axis=1, keepdims=True)

def compute_softmax_loss(X_b, y_one_hot, W):
    """计算Softmax交叉熵损失"""
    m = X_b.shape[0]
    scores = X_b @ W
    probabilities = softmax(scores)
    # 避免log(0)
    probabilities = np.clip(probabilities, 1e-10, 1 - 1e-10)
    loss = -np.sum(y_one_hot * np.log(probabilities)) / m
    return loss

def gradient_descent_softmax(X, y_one_hot, num_classes, learning_rate, num_iterations):
    """
    使用梯度下降训练Softmax Regression模型。
    X: 特征矩阵 (m, n)
    y_one_hot: one-hot编码的标签矩阵 (m, num_classes)
    num_classes: 类别数量
    learning_rate: 学习率
    num_iterations: 迭代次数
    """
    m, n = X.shape
    # 初始化权重矩阵W，额外加一行作为偏置项
    W = np.zeros((n + 1, num_classes))
    # 将偏置项x0=1添加到X
    X_b = np.c_[np.ones(m), X]

    losses = []

    for i in range(num_iterations):
        scores = X_b @ W
        probabilities = softmax(scores)
        gradient = (1 / m) * X_b.T @ (probabilities - y_one_hot)
        W -= learning_rate * gradient
        loss = compute_softmax_loss(X_b, y_one_hot, W)
        losses.append(loss)

        if (i + 1) % 1000 == 0:
            print(f"Iteration {i+1}/{num_iterations}, Loss: {loss:.4f}")
    return W, losses

def predict_softmax_regression(X, W_trained):
    """
    使用训练好的权重进行预测。
    X: 特征矩阵
    W_trained: 训练好的权重矩阵 (包含偏置项)
    返回预测的类别标签
    """
    m = X.shape[0]
    X_b = np.c_[np.ones(m), X] # 添加偏置项
    scores = X_b @ W_trained
    probabilities = softmax(scores)
    return np.argmax(probabilities, axis=1)

def get_probabilities_softmax(X, W_trained):
    """
    获取每个样本对每个类别的预测概率。
    """
    m = X.shape[0]
    X_b = np.c_[np.ones(m), X] # 添加偏置项
    scores = X_b @ W_trained
    return softmax(scores)

# --- 训练Softmax Regression模型 ---
num_classes = len(target_names)
# 将标签y_train转换为one-hot编码
y_train_one_hot = np.eye(num_classes)[y_train]

softmax_learning_rate = 0.1
softmax_num_iterations = 20000 # Softmax通常需要更多迭代

W_softmax, softmax_losses = gradient_descent_softmax(X_train_pca, y_train_one_hot, num_classes,
                                                     softmax_learning_rate, softmax_num_iterations)
print(f"\nCompleted training for Softmax Regression: \n{W_softmax}")

# 对测试集进行预测
y_pred_softmax = predict_softmax_regression(X_test_pca, W_softmax)

# 计算准确率
accuracy_softmax = np.mean(y_pred_softmax == y_test)
print(f"\nSoftmax Regression Test Accuracy: {accuracy_softmax:.4f}")

# --- 绘制Softmax Regression的决策边界 ---
plt.figure(figsize=(7, 6))

x_min, x_max = X_pca[:, 0].min() - 1, X_pca[:, 0].max() + 1
y_min, y_max = X_pca[:, 1].min() - 1, X_pca[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))

# 使用Softmax预测网格点
Z_softmax = predict_softmax_regression(np.c_[xx.ravel(), yy.ravel()], W_softmax)
Z_softmax = Z_softmax.reshape(xx.shape)

# 绘制背景色区域
plt.contourf(xx, yy, Z_softmax, alpha=0.4, cmap=plt.cm.viridis)

# 绘制散点图
for i, target_name in enumerate(target_names):
    plt.scatter(X_pca[y_full == i, 0], X_pca[y_full == i, 1], label=target_name, alpha=0.8,
                cmap=plt.cm.viridis, edgecolor='k', s=20)

plt.title(f'Softmax Regression Decision Boundary, Accuracy: {accuracy_softmax:.4f}')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# --- 定义人造特征点 P1, P2, P3 (在PCA降维后的2D空间中) ---
# 这些坐标是根据Iris数据集的PCA图大致估算的，您可以根据实际绘图调整
P1 = np.array([-3.5, 2.0])  # 远离所有类别，例如在左上角
P2 = np.array([3.0, -1.5])  # 远离所有类别，例如在右下角
P3 = np.array([0.5, 0.5])   # 位于类别之间，例如versicolor和virginica的边界附近

# 将这些点组合成一个数组，方便批量处理
artificial_points = np.array([P1, P2, P3])
point_labels = ['P1', 'P2', 'P3']

print("人工特征点 (PCA坐标):")
for i, p in enumerate(artificial_points):
    print(f"{point_labels[i]}: {p}")

# --- 使用Softmax模型进行预测概率 ---
probabilities_P = get_probabilities_softmax(artificial_points, W_softmax)

print("\n--- Softmax 模型对人工特征点的预测概率 ---")
for i, p_probs in enumerate(probabilities_P):
    print(f"{point_labels[i]}:")
    for j, prob in enumerate(p_probs):
        print(f"  P(Class={target_names[j]}) = {prob:.4f}")
    predicted_class_idx = np.argmax(p_probs)
    print(f"  预测类别: {target_names[predicted_class_idx]}")
    print("-" * 20)


# --- 可视化人工特征点及其预测 ---
plt.figure(figsize=(8, 7))

# 绘制Softmax决策边界作为背景
x_min, x_max = X_pca[:, 0].min() - 1, X_pca[:, 0].max() + 1
y_min, y_max = X_pca[:, 1].min() - 1, X_pca[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))
Z_softmax = predict_softmax_regression(np.c_[xx.ravel(), yy.ravel()], W_softmax)
Z_softmax = Z_softmax.reshape(xx.shape)
plt.contourf(xx, yy, Z_softmax, alpha=0.4, cmap=plt.cm.viridis)

# 绘制原始数据点
for i, target_name in enumerate(target_names):
    plt.scatter(X_pca[y_full == i, 0], X_pca[y_full == i, 1], label=target_name, alpha=0.6, s=30)

# 绘制人工特征点
point_colors = ['red', 'orange', 'purple'] # 为P1, P2, P3选择不同颜色
for i, p in enumerate(artificial_points):
    plt.scatter(p[0], p[1], s=150, marker='*', edgecolor='black', linewidth=1.5,
                label=f'{point_labels[i]}', color=point_colors[i], zorder=10) # zorder确保点在最前面

plt.title('Softmax Decision Boundary with Artificial Points P1, P2, P3')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# 一. Logistic Regression 二分类器 (投票法)
# 1. 决策边界：每个分类器都试图在两个特定类别之间找到一个线性分隔面。例如，“Class 0 vs Class 1”的分类器在 Setosa 和 Versicolor 之间划线。
# 2. 投票法多分类：在多分类决策边界图中，可以看到投票法形成了一个更复杂的非线性边界，尽管每个基础分类器都是线性的。
# 3.准确率：投票法通常能提供一个合理的基线性能。
# 二. Softmax Regression 多分类器
# 1. 决策边界：从图中可以看出，它将特征空间划分为三个区域，每个区域对应一个预测类别。这些边界通常也是线性的（对于线性Softmax）。
# 2. 准确率：与投票法相比，Softmax通常在类别之间界限清晰时表现更好，因为它直接优化多分类问题。
# 3. 损失曲线（可选但推荐）：如果您绘制了损失曲线，可以观察损失是否随着迭代次数的增加而收敛，以判断模型是否训练良好。
# 三. 对“人造”特征点 P1, P2, P3 的预测分析
# 1. P1 和 P2 (分布外特征点)：
# 1.1 理论上，对于远离所有训练数据分布的点，模型可能会给出比较“犹豫”的预测（即所有类别的概率都相对较低或某几个类别概率接近），或者因为模型没有见过类似数据而给出“自信但错误”的预测。
# 1.2 这突出了模型在处理新颖、未见过的数据时的局限性（泛化能力）。
# 2. P3 (分类界面交界点)：
# 2.1 由于它位于决策边界附近，预计其预测概率分布会比较“模糊”，即其所在边界相邻的两个或多个类别的概率会比较接近。
# 2.2 这反映了模型在决策边界附近的不确定性，是分类器固有的特点。
```
